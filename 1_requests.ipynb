### 웹 크롤링 기술

```python
# requests

#  클라이언트(웹 브라우저)
#  ↓(request) ↑(response) 
#         서버

# 회사(AI 부서) : 고양이, 강아지 분류하는 인공지능 모델 개발!
# 회사(웹 부서) : 웹 서비스 개발

# 1. 프론트엔드(이미지 업로드)
# 2. 백엔드(API)
# 3. 백엔드 -> request -> 인공지능 API

# 프론트엔드 -> 백엔드 API -> request -> 인공지능 API
#                                       ↓ 
# 프론트엔드 ← 백엔드 API ← response ← 결과

import requests

response = requests.get("http://news.naver.com")
print(response)

# HTTP는 respnse(status_code : 200, 400, 500)
# 200: 통신 성공 [요청 → 응답 받음!]
# 400: 클라이언트측 ERROR
# 500 : 서버측 ERROR
```

```python
# requests 모듈은 응답으로 해당 URL의 페이지소스 전체 가지고 옴!
print(response.text)
```

```python
# Beautifulsoup
# - requests가 가져온 전체 html 소스 안에서
# 선택자를 이용해 원하는 데이터를 추출
# - 문서를 파싱하여, 트리 구조로 만들어서 특정 태그를 쉽게 찾을 수 있도록 도와주는 라이브러리
from bs4 import BeautifulSoup as bs

soup = bs(response.text, "lxml") # -> python이 html을 읽을 수 있도록 하여 dom트리로 만듦
soup
```

```python
# id 선택자 : #
# class 선택자 : .
# 자식 선택자 : >
# 자손 선택자: 한칸띄기

# Python 컬렉션 타입
# 1. list []
# 2. tuple ()
# 3. dict {key:value} -> JSON Type
# 4. set {}

# bs의 select() -> List Type
data = soup.select("span.Nitem_link_menu")

print("[여러개]",data) # -> list형태로 모든 link_menu 가져옴
print("[1개만]",data[5]) # -> 인덱스 5번 link_menu 만 가져옴
```

```python
# .text는 태그정보 지우기
for i in data:
    print(i.text)
```

```python
# 기존 Python
news_item = []
for i in data:
    news_item.append(i.text)
print(news_item)

# 리스트 컴프리헨슨
news_item = [i.text for i in data]
print(news_item)
```

### 네이버에서 LLM 검색 후 나오는 뉴스 제목 수집하기!

```python
# 1. 데이터 수집 URL 설정
# 네이버 뉴스 LLM 검색 결과 URL
url = "https://search.naver.com/search.naver?where=news&ie=utf8&sm=nws_hty&query=llm"

#2.html 가져오기
response = requests.get(url)
html = response.text

#3. python 형식으로 html 트리 인덱싱(beautifulsoup)
soup = bs(html, "lxml")

#4. 원하는 데이터 select
data = soup.select("a.news_tit")
for i, val in enumerate(data):
    print(f"{i+1}. {val.text}")
```
