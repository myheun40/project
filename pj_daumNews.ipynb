### 1. 다음 뉴스 1건 수집(제목, 날짜, 본문)

```python
import requests as req
from bs4 import BeautifulSoup as bs

#다음 뉴스 1건(제목,날짜, 본문)

url = "https://v.daum.net/v/20240812112013947"
res = req.get(url)
html = res.text

soup = bs(html,"lxml")
title = soup.select("h3.tit_view")[0].text # -> 값이 하나밖에 없어도 리스트 형태로 담기기 때문에 인덱싱 해줘야함
reg_date = soup.select("span.num_date")[0].text
content_list = soup.select("div.article_view p")
# content_list -> 분리 된 뉴스 본문(p) 여러개 List로 지정

content = ""
for i in content_list:
    content+=i.text

print(f"제목: {title}")
print(f"날짜: {reg_date[:12]}") # 리스트 슬라이싱
print(f"본문: {content}")
```

```python
# 단건 뉴스 수집 함수화
def get_news(url: str):
    # 매개변수로 받은 URL의 제목, 날짜, 본문 데이터 수집!
    res = req.get(url)
    soup = bs(res.text,"lxml")

    #제목, 날짜, 본문
    title = soup.select("h3.tit_view")[0].text # -> 값이 하나밖에 없어도 리스트 형태로 담기기 때문에 인덱싱 해줘야함
    reg_date = soup.select("span.num_date")[0].text
    content_list = soup.select("div.article_view p")
    # content_list -> 분리 된 뉴스 본문(p) 여러개 List로 지정
    
    content = ""
    for i in content_list:
        content+=i.text
    
    print(f"제목: {title}")
    print(f"날짜: {reg_date[:12]}") # 리스트 슬라이싱
    print(f"본문: {content}")
```

### 2. 다음 뉴스 리스트(15건) 수집(제목, 날짜, 본문)

```python
url =  "https://news.daum.net/breakingnews/digital"
html = req.get(url).text

soup = bs(html,"lxml")
url_list = soup.select("ul.list_news2 a.link_txt")

# url_list -> 1page(15건) 뉴스 URL

for i, url in enumerate(url_list):
    print(f"{i+1} ===========================================================================================================")
    get_news(url["href"])
```

### 3. 페이지별로 수집(1p~?) 제목, 본문, 날짜 수집

```python
# 1p~10p 수집 코드
# 마지막 페이지 찾기 알고리즘 -->
# 마지막 페이지에는 
# - len(url_list)==0 멈춤!
page=1
count=0
while (True):
    page_url = f"https://news.daum.net/breakingnews/digital?page={page}"
    soup = bs(req.get(page_url).text,"lxml")
    # url_list -> 1page(15건) 뉴스 URL
    url_list = soup.select("ul.list_news2 a.link_txt")

    if len(url_list)==0:
        break
    
    for i, url in enumerate(url_list):
        print(f"{count+1} ===========================================================================================================")
        get_news(url["href"])
        count+=1
    page+=1
```
