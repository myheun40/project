### EDA (탐색적 데이터 분석)

```python
# 직원들의 이직여부 확인
data['Attrition'].value_counts()
```

```python
# 우리 회사의 이직률 확인
# 이직자수 / 전체수 *100
237/(237+1233)*100
# 약 16% 정도의 이직률
```

```python
# 정답테이더 -> Attrition(이직여부) -> 숫자 데이터
# 이직한 사람 -> 1, 이직하지 않은 사람-> 0
# 성별, 나이대에 따른 이직률을 알아보기 위해 데이터 변경
import numpy as np
# np.where(조건, 조건에 부합하는 데이터 변경, 나머지 데이터 변경)
data['Attrition'] = np.where(data['Attrition']=='Yes',1,0)
```

```python
data
```

```python
# 연령데이터 확인
data['Age'].value_counts()
# 연속적인 데이터 카테고리화 하여 데이터의 특징을 부각
```

```python
# 연령데이터 3개의 카테고리로 묶어주기!
# 30대 이하, 30~40, 40대 이상   -> 구간화, 카테고리화
data['Age_gp'] = np.where(data['Age'] <= 30, '30 이하', np.where(data['Age'] < 40, '30-40', '40대 이상'))
```

```python
 data
```

```python
# 연령대별 이직률 확인
# 이직률 = 이직자 / 전체수 * 100
# 30대 이하 이직률 = 30대 이하 이직자(Attrition sum) / 30대 이하 사람 수(Attrition count)
df_gb = data.groupby('Age_gp')['Attrition'].agg(['sum','count'])
df_gb['ratio'] = round(df_gb['sum']/df_gb['count'] *100, 1)
df_gb
# 연령대가 어릴수록 이직률이 높은것을 확인
```

```python
# 성별에 따른 이직률 현황
# 성별 컬럼: 'Gender'
data['Gender'].unique()
df_gender = data.groupby('Gender')['Attrition'].agg(['sum','count'])
df_gender['ratio'] = round(df_gender['sum']/ df_gender['count'] *100, 1)
df_gender
# 여성보다는 남성의 이직률이 조금 더 높은 것을 확인
```

```python
# 부서별 이직률 현황
# Department
data['Department'].unique()  #부서가 몇개 있는지 확인
df_depart = data.groupby('Department')['Attrition'].agg(['sum','count']) 
df_depart['ratio'] = round(df_depart['sum'] / df_depart['count'] *100,1)
df_depart
```

### 가설1 : 업무만족도는 높으나 인간관계로 인한 이직률이 높을 것이다

```python
# 인간관계 만족도와 업무만족도의 상관관계 확인
# 인간관계 만족도 : RelationshipSatisfaction, 업무만족도: JobSatisfaction
# 숫자가 클수록 만족!
data[['RelationshipSatisfaction','JobSatisfaction','Attrition']].head()
```

```python
# 업무 만족도별 이직률 현황
data['JobSatisfaction'].unique()
df = data.groupby('JobSatisfaction')['Attrition'].agg(['sum','count'])
df['ratio'] = round(df['sum'] / df['count']*100 , 1)
df
# 업무만족도가 낮을수록 이직률이 높다!
```

```python
# 인간관계 만족도별 이직률 현황
data['JobSatisfaction'].unique()
df = data.groupby('RelationshipSatisfaction')['Attrition'].agg(['sum','count'])
df['ratio'] = round(df['sum'] / df['count']*100 , 1)
df
# 인간관계 만족도가 낮을수록 이직률이 높음
```

```python
# 인간관계, 업무만족도 두 컬럼의 관계 확인
df = data.groupby(['JobSatisfaction','RelationshipSatisfaction'])['Attrition'].agg(['sum','count'])
df['ratio'] = round(df['sum'] / df['count']*100 , 1)
df
# 결과 해석
# 업무만족도가 높은 직원은 인간관계에 따른 이직률 변동이 적다(영향을 덜받음)
# 업무만족도가 낮은 직원은 이직률에 인간관계 영향력이 큰 것 확인

# 우리의 가설: 업무만족도는 높으나 인간관계로 인한 이직률이 높을것이다
# 우리의 가설이 잘못 설정 -> 가설 수정 -> 결과 확인
# 업무만족도가 이직률에 미치는 영향이 더 높구나! -> 직원들의 엄무만족도 향상을 위해 힘쓸것! 
```

### 가설2: 근속년수 대비 같은 업무를 한 비중이 높다면 이직률이 높을 것이다!

```python
# YearsInCurrentRole: 현재 역할에서 근무한 기간
# YearsAtCompany : 현재까지 근무한 기간
data[['YearsInCurrentRole','YearsAtCompany','Attrition']]
```

```python
# 근속년수 대비 한가지 업종에서 일한 비중
data['Role_Company'] = data['YearsInCurrentRole'] / data['YearsAtCompany']
data['Role_Company'].fillna(0, inplace=True)
data['Role_Company']
```

```python
import seaborn as sns
```

```python
# Role_Company 분포 확인
# distplot : 데이터의 부포를 확인하는데 주로 사용됨. 
# 히스토그램과 커널밀도추정을 함께 그려서 데이터의 분포를 확인
sns.distplot(data['Role_Company'])
```

```python
# 카테고리화
# np.where()
# '0.3이하', '0.3~0.6', '0.6 초과'
data['Role_Company_gp'] = np.where(data['Role_Company'] <= 0.3, '0.3 이하', np.where(data['Role_Company'] <= 0.6, '0.3~0.6','0.6 초과'))
data[['Role_Company','Role_Company_gp']]
```

```python
# 근속년수 대비 같은 일을 오래 한 사람들의 이직률
df = data.groupby('Role_Company_gp')['Attrition'].agg(['sum','count'])
df['ratio'] = round(df['sum'] / df['count']*100 , 1)
df.loc[['0.3 이하','0.3~0.6','0.6 초과']]
# 근속년수 대비 같은 일을 짧게 한 사람일 수록 이직률이 높은 것!
# 가설 수정 -> 근속년수 대비 한업무를 길게 할 수 있도록 권장
```

### 가설3: 야근을 많이 해도 급여인상률이 높다면 이직률이 낮을 것이다!

```python
# 야근 여부에 따른 이직률 현황
# OverTime
df = data.groupby('OverTime')['Attrition'].agg(['sum','count'])
df['ratio'] = round(df['sum'] / df['count']*100 , 1)
df
# 야근을 많이 한사람은 이직률이 높음
```

```python
# 연봉인상률에 따른 이직률 현황
# PercentSalaryHike
df = data.groupby('PercentSalaryHike')['Attrition'].agg(['sum','count'])
df['ratio'] = round(df['sum'] / df['count']*100 , 1)
df
```

```python
# 야근과 연봉인상률의 상관관계에 따른 이직률
df = data.groupby(['OverTime','PercentSalaryHike'])['Attrition'].agg(['sum','count'])
df['ratio'] = round(df['sum'] / df['count']*100 , 1)
df
```

### 모델선택 및 하이퍼 파라미터 조절
- 데이터 타입별로 구분(숫자/ 문자)
- 문자형태의 데이터는 인코딩 (숫자형태로 변경)
- 데이터 정리(불필요한 컬럼 삭제)
- X,y 분리
- train, test 분리

```python
data
```

```python
# 데이터 타입에 따른 컬럼 분리
for i in data.columns:
    print(data[i].dtype, data[i].dtype == 'O')
```

```python
# 숫자타입의 컬럼과 문자 타입의 컬럼을 각각 리스트에 저장
numeric_list = []
categorical_list = []

for i in data.columns:
    if data[i].dtype == 'O':
        categorical_list.append(i)
    else:
        numeric_list.append(i)   
```

```python
# 모델링을 위한 데이터 분리
# 문제(X), 정답(y)
# 가설 확인을 위해 만들 컬럼들은 삭제
X = data.drop(['Attrition','Age_gp','Role_Company','Role_Company_gp'], axis = 1)
y = data['Attrition']
```

```python
# 레이블 인코딩 - 
# BusinessTravel : 출장빈도 -> Non-Travel, Travel_Rarely, Travel_Frequently	
# 출장이 많을수록 이직률에 영향을 많이 미치지 않을까?
# 출장이 잦은 데이터에는 높은 숫자를 매칭

# 딕셔너리를 통해서 각 데이터에 숫자 매칭
BT_dict = {'Non-Travel':0, 'Travel_Rarely':1, 'Travel_Frequently':2}

# map() : 값을 매핑해주는 함수 같은 키값을 밸류로 변경
X['BusinessTravel'] = X['BusinessTravel'].map(BT_dict)
```

```python
X['BusinessTravel']
```

```python
# 나머지 컬럼은 원핫인코딩 진행!
# 원핫인코딩 - 우선순위가 없을 때 사용
X_one_hot = pd.get_dummies(X, dtype = int)
```

```python
# 데이터 분리 도구 불러오기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_one_hot, y, test_size=0.3, random_state=1)
```

```python
#데이터 크기확인
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### 모델링
- 모델 객체생성
- 모델 학습
- 모델 평가

```python
# 모델 객체생성 LogisticRegression
from sklearn.linear_model import LogisticRegression
logi_model = LogisticRegression()  #객체생성 시 ()소괄호 사용해주기
```

```python
# 모델 학습
# 모델명.fit(학습용문제, 학습용정답)
logi_model.fit(X_train, y_train)
```

```python
# 모델 평가
# 모델명.score(테스트용 문제, 테스트용 정답)
logi_model.score(X_test, y_test)
# 정확도 (accuracy )
```

### 교차검증
- 모델의 일반화 성능평가
- 일반화 성능 측정 방법: 모델이 데이터에 대해 얼마나 잘 맞추는 지 한번에 평가하는 것보다 여러번 나누어 평가하는 것이 신뢰성이 높다.
- 방법 : (훈련용데이터 내에서) 훈련세트와 테스트세트로 나누어 여러겹 학습 및 평가
- 최종 평가전에 평균 정확도 확인
- 평균적인 성능을 확인하면, 한번 평가하는 것 보다 일반화에 대한 판단을 할 수 있다.
- 과대적합인지 아닌지를 확인하는 과정으로써, 과대적합을 제어하는 것은 아님!
- 일반화(어떤 데이터셋이 들어와도 평균적인 성능을 보임) 정도를 확인


```python
# 교차검증
from sklearn.model_selection import cross_val_score #교차검증 도구
# cross_val_score(모델명, 학습용문제, 학습용정답, cv=5)
result = cross_val_score(logi_model,X_train, y_train, cv = 5)
result.mean()
```

# Support vector machine(SVM)

```python
from sklearn.svm import LinearSVC  #분류모델
# 모델 객체생성
svm_model = LinearSVC()
```

```python
# 모델 학습
# model.fit(학습용문제, 학습용 정답)
svm_model.fit(X_train, y_train)
```

```python
# 모델 평가
# model.score(테스트용 문제, 테스트용 정답)
svm_model.score(X_test, y_test)
# 분류평가지표: 정확도 accuracy -> 0~1 -> 1에 가까울 수록 좋음
```

```python
# 교차검증
result = cross_val_score(svm_model,X_train, y_train, cv = 5)
result
```

### 평가지표
- 정확도(accuracy) : 전제 데이터중에서 정답을 맞춘 비율
- 재현율(recall) : 실제양성중에서 예측 양성의 비율
- 정밀도(pricision) : 예측양성중에서 실제양성의 비율
- f-1 score : 재현율과 정밀도의 조화 평균
- classification_report

```python
from sklearn.metrics import classification_report
# classification_report(실제정답, 예측정답)
# 얘측 -> 학습모델.predict(테스트용문제)

#logi_logi_model 예측결과
logi_pre = logi_model.predict(X_test)
print(classification_report(y_test, logi_pre))

# support: 각 클래스의 실제 샘플 수
# class 0: 대부분 정확하게 맞추고 있음
# class 1: 매우 낮은 결과값 확인 -> 모델이 class가 1인 경우에는 전혀 예측을 하지 못하고 있음
# accuracy 0.84로서 보통수준이지만, class 1인 경우에 많이 오답을 예측 -> 평군점수가 낮아진것 같음
# 결론, 모델은 class 0에 대해서는 잘 예측하였지만, class 1에 대해서는 예측성이 낮음.
# 이는 데이터클래스의 불균형문제를 해결하거나, 모델의 성능을 개선해야할 필요가 있음.
```
