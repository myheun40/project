## EDA(탐색적 데이터 분석)

```python
# 직원들의 이직여부 확인
data['Attrition'].value_counts()
```

```python
# 우리 회사의 이직률 확인
237/(237+1233)
# 약 16% 정도의 이직률
```

```python
# 정답데이터 -> Attrition(이직여부) -> 숫자 데이터
# 이직한 사람 -> 1, 이직하지 않은 사람 -> 0
import numpy as np
# np.where(조건, 조건에 부합하는 데이터 변경, 나머지 데이터 변경)
data['Attrition']=np.where(data['Attrition']=="Yes", 1, 0)
```

```python
# 연령데이터 확인
data['Age'].value_counts()
```

```python
# 연력데이터 3개의 카테고리로 묶어주기!
# 30대 이하, 30-40, 40대 이상
data['Age_gp'] = np.where(data['Age']<=30,'30 이하', np.where(data['Age']<=40, '30-40', '40대 이상'))
```

```python
data
```

```python
# 연령대별 이직률 확인
# 이직률 = 이직자 / 전체수 * 100
# 30대 이하 이직률 = 30대 이하 이직자 (Attrition sum) / 30대 이하 사람수 (Attrition count) 
df_gb = data.groupby('Age_gp')['Attrition'].agg(['sum','count'])
df_gb['ratio'] = round(df_gb['sum']/df_gb['count']*100, 1)
df_gb
# 연령대가 낮을수록 이직률이 높은 것을 확인
```

```python
# 성별에 따른 이직률 현황
# 성별 컬럼: 'Gender'
df_gender = data.groupby(by='Gender')['Attrition'].agg(['sum','count'])
df_gender['ratio']=round(df_gender['sum']/df_gender['count']*100, 1)
df_gender
# 여성보다 남성의 이직률이 더 높음
```

```python
# 부서별 이직률 현황
# Department
df_dp=data.groupby('Department')['Attrition'].agg(['sum','count'])
df_dp['ratio']=round(df_dp['sum']/df_dp['count']*100,1)
df_dp
# 판매 부서의 이직률이 더 높다
```

```python
# 인간관계 만족도별 이직률현황
df = data.groupby('RelationshipSatisfaction')['Attrition'].agg(['sum','count'])
df['ratio']=round(df['sum']/df['count']*100,1)                                                                           
df
# 업무만족도가 낮을수록 이직률이 높다
```

```python
# 인간관계, 업무만족도 두 컬럼의 관계 확인
df = data.groupby(['JobSatisfaction','RelationshipSatisfaction'])['Attrition'].agg(['sum','count'])
df['ratio']=round(df['sum']/df['count']*100,1)
df
# 결과해석
# 업무만족도가 높은 직원은 인간관계에 따른 이직률 변동이 적다(영향을 덜 받음)
# 업무만족도가 낮은 직원은 이직률에 인간관계 영향력이 큰 것 확인

# 우리의가설: 업무만족도는 높으나 인간관계로 인한 이직률이 높을 것이다
# 우리의 가설이 잘못 설정됨 -> 가설 수정 -> 결과 확인
# 업무만족도가 이직률에 미치는 영향이 더 높구나! -> 직원들의 업무 만족도 향상을 위해 힘쓸 것!
```

### 가설2) 근속년수 대비 같은 업무를 한 비중이 높다면 이직률이 높을 것이다?

```python
# YearsInCurrentRole: 현재 역할에서 근무한 기간
# YearsAtCompany : 현재까지 근무한 기간
data[['YearsInCurrentRole','YearsAtCompany','Attrition']]
```

```python
# 근속년수 대비 한가지 업종에서 일한 비중
data['Role_Company']=data['YearsInCurrentRole']/data['YearsAtCompany']
data['Role_Company'].fillna(0, inplace=True)
data['Role_Company']
```

```python
import seaborn as sns
```

```python
# Role_Company 분포 확인
# distplot : 데이터의 분포를 확인하는데 주로 사용됨
# 히스토그램과 커널밀도추정을 함께 그려서 데이터의 분포를 확인
sns.distplot(data['Role_Company'])
```

```python
# 카테고리화~
# np.where()
# '0.3이하', '0.3-0.6', '0.6 초과'
# Role_Company_gp
data['Role_Company_gp'] = np.where(data['Role_Company']<=0.3, '0.3이하', np.where(data['Role_Company']<=0.6, '0.3-0.6', '0.6초과'))
data[['Role_Company','Role_Company_gp']]
```

```python
# 근속년수 대비 같은일을 오래 한 사람들의 이직률 파악
df = data.groupby('Role_Company_gp')['Attrition'].agg(['sum','count'])
df['ratio']=round(df['sum']/df['count']*100,1)
df.loc[['0.3이하','0.3-0.6','0.6초과']]
# 근속 년수 대비 같은 일을 짧게 한 사람일 수록 이직률이 높은 것!
# 가설 수정 -> 근속년수 대비 한 업무를 길게 할 수 있도록 권장
```

### 가설3) 야근을 많이 해도 급여인상률이 높다면 이직률이 낮을 것이다

```python
data['OverTime']
```

```python
# 야근여부에 따른 이직률 현황
# OverTime
df_overtime = data.groupby('OverTime')['Attrition'].agg(['sum','count'])
df_overtime['ratio']=round(df_overtime['sum']/df_overtime['count']*100,1)
df_overtime
```

```python
# 연봉인상률에 따른 이직률 현황
# PercentSalaryHike
df_p=data.groupby('PercentSalaryHike')['Attrition'].agg(['sum','count'])
df_p['ratio'] = round(df_p['sum']/df_p['count']*100,1)
df_p
```

```python
# 야근과 연봉인상률의 상관관계에 따른 이직률
df=data.groupby(['OverTime','PercentSalaryHike'])['Attrition'].agg(['sum','count'])
df['ratio'] = round(df['sum']/df['count']*100,1)
df
```

```python
# 야근 No 직원들의 평균 이직률 확인
temp = df.reset_index()
temp[temp['OverTime']=="No"]['ratio'].mean()
# 야근을 하지 않는 직원의 연봉인상률
```

```python
# 야근 Yes 직원들의 평균 이직률 확인
temp[temp['OverTime']=="Yes"]['ratio'].mean()
# 결과 -> 야근을 할수록 이직률이 높은 것을 확인
```

```python
# 야근을 Yes하는 직원의 데이터 분석
over_yes = temp[temp['OverTime']=="Yes"]
over_yes
# 시각화(막대그래프, 연봉인상률(X), 이직률(y))
import matplotlib.pyplot as plt
plt.bar(over_yes['PercentSalaryHike'], over_yes['ratio'])
plt.show()
# 연봉인상률 20%까지는 이직률이 조금은 감소하는 듯 싶으나
# 21% 이상부터는 이직률이 증가(핵심인재의 이탈)
# 연봉인상률이 높은 직원은 야근시간을 줄이는 것이 핵심인재의 이탈을 방지하는 요소 
```

### 모델선택 및 하이퍼 파라미터 조절
- 데이터 타입별로 구분(숫자/문자)
- 문자형태의 데이터는 인코딩 (숫자형태로 변경)
- 데이터 정리 (불필요한 컬럼 삭제)
- X,y 분리
- train, test 데이터 분리

```python
data.info()
```

```python
# 데이터 타입에 따른 컬럼 분리
for i in data.columns:
    print(data[i].dtype, data[i].dtype== 'O')
```

```python
# 숫자 타입의 컬럼과 문자타입의 컬럼을 각각 리스트에 저장
numeric_list = []
categorical_list = []

for i in data.columns:
    if data[i].dtype =='O':
        categorical_list.append(i)
    else:
        numeric_list.append(i)
```

```python
print(categorical_list)
```

```python
numeric_list
```

```python
# 모델링을 위한 데이터 분리
# 문제(X), 정답(y)
# 가설 확인을 위해 만들 컬럼들은 삭제
X = data.drop(['Attrition'], axis=1)
y = data['Attrition']
```

```python
# 레이블 인코딩
# BusinessTravel : 출장빈도 ->  Non-Travel, Travel_Rarely, Travel_Frequently
# 출장이 많을 수록 이직률에 영향을 많이 미치지 않을까?
# 출장이 잦은 데이터에는 높은 숫자를 매칭

# 딕셔너리를 통해서 각 데이터에 숫자 매칭
BT_dict = {'Non-Travel':0, 'Travel_Rarely':1, 'Travel_Frequently':2}

# map(): 값을 맵핑해주는 함수 같은 키값을 밸류로 변경
X['BusinessTravel']=X['BusinessTravel'].map(BT_dict)
```

```python
X['BusinessTravel']
```

```python
# 나머지 컬럼은 원핫인코딩 진행!
X_one_hot = pd.get_dummies(X, dtype=int)
```

```python
# 데이터 분리 도구 불러오기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_one_hot, y, test_size=0.3, random_state=1)
```

```python
# 데이터 크기확인
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### 모델링
- 모델 객체 생성
- 모델 학습
- 모델 평가

```python
# 모델 객체 생성 LogisticRegression
from sklearn.linear_model import LogisticRegression
logi_model = LogisticRegression()
```

```python
# 모델 학습
# 모델명.fit(학습용문제, 학습용정답)
logi_model.fit(X_train, y_train)
```

```python
# 평가
# 모델명.score(테스트용 문제, 테스트용 정답)
logi_model.score(X_test, y_test)
# 정확도 accuracy
```
