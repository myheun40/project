```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = "https://www.melon.com/chart/index.htm"

# 200번대 : 성공
# 400번대 : 클라리언트 측 잘못 요청(프런트->백으로 잘못 요청)
# 500번대 : 백엔드(서버)쪽 잘못 -- 서버 터졌을 때 (다시 실행)

# 404: not found(URL 찾을 수 없음)
# 405: 페이지는 찾았으나 알맞은 요청이 아닌 경우 -- ex) post로 요청해야하는데 get으로 요청한 경우 [method not allowed]
# 406: 통신 실패

# HTTP 프로토콜
# 1. 일반 헤더
# 2. 요청/응답 헤더
# 3. 엔티티 헤더
# 4. 

# 웹사이트로 접속한 척
user_agent = {
    "USER-Agent" : "Mozilla/5.0 (Windows NT 10.0; win64; x64) AppleWebkit/537.36(KHTML, like Gecok) Chrome/122.0.0.0 Safari/537.36"
}

res = requests.get(url, headers=user_agent)
print(res)
```

```python
soup = bs(res.text, "lxml")
soup
```

```python
title_list = soup.select("div.ellipsis.rank01 > span > a")

title_100 = []
for i, title in enumerate(title_list):
    print(f"{i+1}.{title.text}")
    title_100.append(title.text)
```

```python
print(len(title_100))
```

```python
singer_list = soup.select("div.rank02 > span")

singer_100 = []
for i,singer in enumerate(singer_list):
    print(f"{i+1}. {singer.text}")
    singer_100.append(singer.text)
```

```python
i=1
for title, singer in zip(title_100, singer_100):
    print(f"{i}. {title} - {singer}")
    i+=1
```

```python
# LIST(제목)
# LIST(가수)

# -> 표 형태로 변형(DataFrame)

import pandas as pd

dict={
    "노래 제목": title_100,
    "가수": singer_100
}

melon = pd.DataFrame(dict)
melon
```

```python
melon.to_csv("melon_top_100.csv", encoding="utf-8")
```
